# Qn 1(b)

The Diffusion Equation is given as
\begin{equation}
\dfrac{\partial c}{\partial t} = D\dfrac{\partial^2 c}{\partial x^2}
\end{equation}

where $\tau = Dt$ is a time-like variable rescaled by the <em>diffusion constant</em> $D$.

\begin{equation}
\dfrac{\partial c}{\partial \tau} = \dfrac{\partial^2 c}{\partial x^2} \\
\end{equation}

Approximate the derivative using the central difference theorem

\begin{equation}
\dfrac{\partial c}{\partial \tau} \approx \dfrac{c(u,\tau+1)-c(u,\tau)}{\Delta\tau}\
\dfrac{\partial c}{\partial u} \approx \dfrac{c(u+1,\tau)-c(u,\tau)}{\Delta u}\\
\dfrac{\partial^2 c}{\partial u^2} \approx \dfrac{c(u+1,\tau)-2c(u,\tau)+c(u-1,\tau)}{\Delta u^2}\\
\end{equation}

We then make use of this and substitute to the equation in 1(a)

\begin{equation}
\dfrac{\partial^2 c}{\partial x^2} \approx \ 
\
(1-u^2)^2 
\dfrac{c(u+1,\tau)-2c(u,\tau)+c(u-1,\tau)}{\Delta u^2}\ - 2u(1-u^2)\dfrac{c(u+1,\tau)-c(u,\tau)}{\Delta u}\
\end{equation}

Then subbing in to the diffusion equation we get:


\begin{equation}
\dfrac{c(u,\tau+1)-c(u,\tau)}{\Delta\tau}\ \approx
\
(1-u^2)^2 
\dfrac{c(u+1,\tau)-2c(u,\tau)+c(u-1,\tau)}{\Delta u^2}\ - 2u(1-u^2)\dfrac{c(u+1,\tau)-c(u,\tau)}{\Delta u}\
\end{equation}
